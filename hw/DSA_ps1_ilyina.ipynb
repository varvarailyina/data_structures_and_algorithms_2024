#### 3. Convert the following fractions into 8-bit floating point (1 sign bit, 3 exponent bits and 4 mantissa bits). If that isn’t possible exactly, explain why, and show how inaccurate the encoding would be.

* 3 $\frac{1}{4}$

__Solution:__ \
First, I detect that the sign bit is 0, since both the fraction 3 $\frac{1}{4}$ and the decimal integer 3.25 are positive. I convert the digits before the decimal point (3 in this case) to `11` by continuously dividing by 2 and keeping track of the remainders. Then, I convert the digits after the decimal point (0.25 in this case) to `01` by continuously multiplying by 2 and keeping track of the digit before the decimal. \

In fixed point notation, the result can be shown as `11.01`, which is then converted to 1.101 * 2<sup>1</sup>. Here, the exponent is `127 + 1 = 128`, which when converted to binary is `10000000`. The mantissa is the part after the decimal point, namely `101`, and the sign is `0` because 3.25 is positive.

Combined and written as an 8-bit floating point with 1 sign bit, 3 exponent bit and 4 mantissa bits, some parts of the exponent must be dropped and a zero must be added to the mantissa.

The final result is `01001010`.

* Program counter A0:
    * Fetch: 24AA
    * Decode: LOAD the register 4 with the bit pattern AA
    * Execute: The content of AA is loaded into R4, which is B4 with the bit pattern `10110100`
* Program counter A2
    * Fetch: 33AA
    * Decode: STORE the bit pattern found in register 3 in the memory cell AA
    * Execute: The content of R3 is 33 with the bit pattern `00100001`, which is now stored in AA instead of B4
* Program counter A4
    * Fetch: A303
    * Decode: ROTATE the bit pattern in register 3 on bit to the right 3 times
    * Execute: The bit pattern in R3 is 33 with the bit pattern `00100001`, which is rotated to generate `00100100`
* Program counter A6
    * Fetch: 8234
    * Decode: AND the bit pattern in R3 (`00100100`) and R4 (`10110100`) and place the result in R2
    * Execute: The result of `00100100` AND `10110100` is `00100100`, which is then stored in R2
* Program counter A8

#### 6. Write a Python function that checks if an input number is a palindrome number (a number that reads the same forward and reverse, e.g. 151) and returns True or False; this function should raise an error if the input is not an integer.

````python
## define function
def is_palindrome(num):

    # check if input is an integer
    if not isinstance(num, int):
        raise TypeError("Try again! The input must be an integer.")

    # convert integer into string
    num_str = str(num)

    if num_str == num_str[::-1]:
        print("True")
    else:
        print("False")
````

````python
## test function
num = 1000
is_palindrome(num)
````

````python
num = 1001
is_palindrome(num)
````
